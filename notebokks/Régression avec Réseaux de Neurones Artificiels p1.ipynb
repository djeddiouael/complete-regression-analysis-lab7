```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
np.random.seed(42)
X = np.linspace(-10, 40, 200).reshape(-1,1)
y = 5*np.sin(0.2*X) + 0.5*X + np.random.normal(0, 1, X.shape)
plt.scatter(X, y, color='red')
plt.xlabel('Température (°C)')
plt.ylabel('Consommation énergétique (kWh)')
plt.title('Relation non linéaire')
plt.show()
```

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Generate sample data (REPLACE THIS WITH YOUR ACTUAL DATA)
np.random.seed(42)
x = np.linspace(0, 10, 1000)  # Input features
y = 2 * x + 1 + np.random.normal(0, 1, 1000)  # Target values with noise

# Create the model
model = Sequential([
    Dense(32, activation='relu', input_shape=(1,)),
    Dense(32, activation='relu'),
    Dense(1, activation='linear')
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Train the model
history = model.fit(x, y, validation_split=0.2, epochs=200, verbose=0)

print("Training completed successfully!")
print(f"Final training loss: {history.history['loss'][-1]:.4f}")
print(f"Final validation loss: {history.history['val_loss'][-1]:.4f}")
```

```python
from sklearn.metrics import mean_squared_error, r2_score
y_pred = model.predict(x)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print("MSE:", mse)
print("R2:", r2)
```
